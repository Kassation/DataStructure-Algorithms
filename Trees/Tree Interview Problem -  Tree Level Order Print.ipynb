### Problem

Given a binary tree of integers, print it in level order. The output will contain space between the numbers in the same level, and new line between different levels. For example, if the tree is:

The output should be:
```
1 
2 3 
4 5 6
```


### Solution

````python
class Node:
    def __init__(self, val = None):
        self.left = None
        self.right = None
        self.val = val
````

````python
import collections

def levelOrderPrint(tree):
    if not tree:
        return
    
    #We will use a queue in breadth first search
    #We push the root node into the queue first.
    nodes = collection.deque([tree])
    
    #Current level count indicate how many nodes should be 
    #printed at this level before printing a new line.
    #We discrement is every time we pop an element from the
    #queue and print it.
    currentCount = 1
    
    #Next level count contains the number of nodes in the next
    #level, which will become the current level count after being 
    #printing a new line.
    NextCount = 0
    
    while len(nodes) != 0:
        
        #At each iteration we pop a node from the begining of the queue
        #and push its children to the end of the queue.
        #Once we pop a node, we print its value and space.
        currentNode = nodes.popleft()
        currentCount -= 1
        print(currentNode.val)
        
        if currentNode.left:
            nodes.append(currentNode.left)
            nextCount += 1
            
        if currentNode.right:
            nodes.append(currentNode.right)
            nextCount += 1
        
        #Once the current count reaches zero we print a new line.
        if currentCount == 0:
            print('\n')
            currentCount, nextCount = nextCount, currentCount
            
````


The time complexity of this solution is O(N), which is the number of nodes in the tree, so it’s optimal. Because we should visit each node at least once. The space complexity depends on maximum size of the queue at any point, which is the most number of nodes at one level. The worst case occurs when the tree is a complete binary tree, which means each level is completely filled with maximum number of nodes possible. In this case, the most number of nodes appear at the last level, which is (N+1)/2 where N is the total number of nodes. So the space complexity is also O(N). Which is also optimal while using a queue.
